## A 360 Analysis of Chicago’s 5 year L-Ridership to O’Hare and Midway 

The goal of this project was to use the CTA L-ridership data, O'Hare and Midway Flight and passenger volume data and the Taxi and Transportation Network Providers (TNPS like Uber, Lyft) data to better understand:

1. Total CTA L-ridership trend from 2016 to 2019  for O’Hare and Midway Airports
2.	Percent difference in L- ridership between 2016 and 2019 
3.	Population trend near the community areas around O’Hare and Midway airports from 2010 to 2017
4.	L- Ridership for O’Hare and Midway Airport  for 2019 by month 
5.	Trends in taxi and Transportation Network Ridership(TNP) rides for drop off and pick up at O’Hare and Midway Airports for 2019 by month

## Insights Generated

####  A. Overall CTA L ridership from 2016 to 2019 for O’Hare and Midway Airports:
**The overall ridership declined between 2016 to 2019 for both airports:**
1. O’Hare Ridership declined by 5% between 2016 and 2019
2. Midway Ridership declined by 12% between 2016 and 2019

#### B. O'Hare and Midway CCA Area Population change from 2000 to 2017 and from 2010 to 2017:

**The population in both the areas saw an increase from 2010 to 2017.**
1. CCA 76 for O’Hare - The population saw an increase of 12.3% fro 2010 to 2017
2. CCA 56 for Midway Airport - The population saw an increase of 5.5% from 2010 to 2017

#### C. Flight volumes for O'Hare and Midway Airports from 2016 to 2019
**O'Hare saw an increase in flight volume from 2016 to 2019 while Midway saw a decrease in filght volume during that period.**
1. O'Hare experienced a 32.2% increase in flight volume from 2016 to 2019
2. Midway experienced a 8.2% decrease in flight volume from 2016 to 2019

#### D. Passenger volumes for O'Hare and Midway Airports from 2016 to 2019
**O'Hare saw an increase in passenger volume from 2016 to 2019 while Midway saw a decrease in passenger volume during that period.**
1. O'Hare experienced a 8.6% increase in passenger volume(only travelers) from 2016 to 2019
2. Midway experienced a 8.1% decline in passenger volume(only travelers) from 2016 to 2019

#### E. O'Hare Airport Transportation Analysis for 2019: Impact of introduction of Uber, Lyft and Ride Share Companies (TNP) on L and Taxi Ridership
**Introduction of TNP Companies lead to them taking up the bulk share of the the Riderships.Data indicates that O'Hare passengers predominantly use TNP to go to and from O'Hare Airport**
1. In 2019, TNP Rides made up 68-75% of the combined CTA L and TNP rides
2. In 2019, TNP Pickups made up 69 to 76% of the combined Taxi and TNP pickups
3. In 2019, TNP Dropoffs made up 81-98% of the combined TNP and Taxi dropoffs

#### E. Midway Airport Transportation Analysis for 2019: Impact of introduction of Uber, Lyft and Ride Share Companies (TNP) on L and Taxi Ridership
**Introduction of TNP Companies lead to them taking up the bulk share of the the Riderships.Data indicates that Midway passengers predominantly use TNP to go to and from Midway Airport**
1. In 2019, TNP Rides made up 51- 57% of the combined CTA L and TNP rides. This was lower than that observed for O'Hare airport
2. In 2019, Midway TNP pickups made up 81-84% of the combined TNP and Taxi pickups
3. In 2019, Midway TNP Dropoffs made up 83-95% of the combined TNP and Taxi dropoffs


## Key Takeaways

1. L -ridership to both the airports has seen a steady decline from 2016 to 2019
2. Even though the population in both the community areas which house O'Hare and Midway airports have increased from 2010 to 2017, this increase has not led to increased L-ridership
3. O'Hare airport saw an increase in flight volume from 2016 to 2019 by (32%). Meanwhile, Midway Airport experienced a decline in flight volume during that same period (-8.2%).
4. O'Hare airport saw an increase in passenger volume from 2016 to 2019 by (8.6%). Meanwhile, Midway Airport experienced a decline in passenger volume during that same period (-8.1%).
5. Introduction of TNP rides at both airports impacted the L-ridership as well as the taxi ridership. Both saw a decrease in ridership during the period evaluated(2019).

### Import the CTA L-ridership data and the CTA L-map stops data into padas dataframe

```python
#Get the Data 1 and 2 Files Use API to connect with the portal and get with Data 1 and 2

# DataFrame 1: CTA L-ridership data
#Get the entire ridership data from 2001 to 2020(Q1)

l_rides = pd.read_csv("https://data.cityofchicago.org/resource/t2rn-p8d7.csv?%24limit=350000")
l_rides.head()
```

### CTA L-Ridership data - create new columns for month and year

```python
# Add date_time, year and month columns to the data set

l_rides['date_time'] = pd.to_datetime(l_rides['month_beginning'])
l_rides["month"] = l_rides[
        "date_time"
].dt.month
l_rides["year"] = l_rides[
        "date_time"
].dt.year
l_rides.head()


# getting 2014 to 2020 ridership data

l_2014_2020_rides = l_rides[(l_rides['year'] > 2013) & (l_rides['year'] <= 2020)]
l_2014_2020_rides.shape
#l_2014_2020_rides.head()
```

### Import the CTA L-map stops data

```python
l_map_stops = pd.read_csv('https://data.cityofchicago.org/resource/8pix-ypme.csv',sep=',')
l_map_stops
```

```python
l_2014_2020_rides[l_2014_2020_rides['stationame'].isin(['Randolph/Wabash', 'Washington/Wabash'])].shape
```

```python
# There are duplicate values for map ID column since it counts the stops for trains going in both directions.
# There are a total of 143 unique stop_ids
# We need to keep only one row and drop the duplicate row.

l_map_stops['map_id'].unique().size
```

### Add the x and y coordinates for each L stop 
### Drop the duplicate rows

```python
#Use the function to create the x and y coordinates for each l-stop
# Add the co-ordinates to each L-stop
# Create a xycoord function to convert the 
def xycoord(Coords):
    Coordinates = literal_eval(Coords)
    lat = Coordinates[0]
    lon = Coordinates[1]

    r_major = 6378137.000
    x = r_major * math.radians(lon)
    scale = x/lon
    y = 180.0/math.pi * math.log(math.tan(math.pi/4.0 + 
        lat * (math.pi/180.0)/2.0)) * scale
    return (x, y)

l_map_stops['coords_x'] = l_map_stops['location'].apply(lambda x: xycoord(x)[0])
l_map_stops['coords_y'] = l_map_stops['location'].apply(lambda x: xycoord(x)[1])
l_map_stops[['location', 'coords_x', 'coords_y']].head()

# Drop duplicates since each stop is for going to and going from directions. ALl other data stays same
# Verify that there are only 143 rows

l_map_stops.drop_duplicates(subset='map_id', keep="last", inplace=True)
l_map_stops.groupby([l_map_stops['map_id']]).count()
```

### Change the column name 'map_id' in L-map Stops dataset to 'station_id' 
### This will allow merging of the two data sets on the station_id column values

```python
# Change the column name "map_id" to match the name in the l_2014_2020_rides column name which is "station_id" 
#This will allow us to merge DataFrames 1 and 2 
#The new merged dataframe wnow contains the line infor added for each station.

l_map_stops.rename(columns = {'map_id':'station_id'}, inplace = True)
l_map_stops.head()
```

### Merge (Left Join) the CTA L-Rides dataset and the L-Map Stops dataset on the column station_id


```python
# Merge the DataFrames 1 and 2 together on the map_id (data) 
# This gives us the ridership across all the L-stops between 2014 and 2020.
# Left Joined the ridership data with the L-stop data so that all the ridership rows are included.

merged_rides_lines = pd.merge(left=l_2014_2020_rides, right=l_map_stops, how='left', left_on='station_id', right_on='station_id')

# merged_rides_lines.info()

# There are 110 null values
merged_rides_lines.shape
```

### Replace na values with 0 for the stop_id feature and with 'noname' for stop_name, direction_id, station_name and station_descriptive_name


```python
#Filled the na rows and columns with their corresponding values as in the ridership datafile

merged_rides_lines.stop_id = merged_rides_lines.stop_id.fillna(0)
merged_rides_lines.stop_name = merged_rides_lines.stop_name.fillna('noname')
merged_rides_lines.direction_id = merged_rides_lines.direction_id.fillna('noname')
merged_rides_lines.station_name = merged_rides_lines.station_name.fillna('noname')
merged_rides_lines.station_descriptive_name = merged_rides_lines.station_descriptive_name.fillna('noname')
```

### Replacing the nan values for the line type feature with True for g, pink, orange, brown and purple lines
### Replacing the nan values for the line type feature with False for red, blue and pepx lines

```python
# Check for Null Values:
# The following fetaures have NANs: stop_id, stop_name, direction_id, station_name
# The values for the ada feature will be 'nodata'
# The two L stops which were closed in 2017 were Randolph/Wabash and Madison/Wabash. 
# These two stops therefore have nan values and they will be replaced with True (for the lines which had these stops)
# Green, Pink, Orange, Brown, Purple and False for Blue, Red, PEXP and Yellow


merged_rides_lines.ada = merged_rides_lines.ada.fillna('nodata')
merged_rides_lines.pnk = merged_rides_lines.pnk.fillna(True)
merged_rides_lines.brn = merged_rides_lines.brn.fillna(True)
merged_rides_lines.g = merged_rides_lines.g.fillna(True)
merged_rides_lines.o = merged_rides_lines.o.fillna(True)
merged_rides_lines.p = merged_rides_lines.p.fillna(True)
merged_rides_lines.red = merged_rides_lines.red.fillna(False)
merged_rides_lines.blue = merged_rides_lines.blue.fillna(False)
merged_rides_lines.pexp = merged_rides_lines.pexp.fillna(False)
merged_rides_lines.y = merged_rides_lines.y.fillna(False)

merged_rides_lines.info()
```
